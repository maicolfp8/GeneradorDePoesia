```python
!pip install openai gradio gtts pydub requests
!pip install openai==0.28.0

import requests
import openai
import gradio as gr
from gtts import gTTS
from pydub import AudioSegment
import os

# Configurar la API Key
openai.api_key = "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"

# Ruta al archivo de m√∫sica de fondo que ya tienes
musica_path=   "/content/musica_fondo.mp3" # ruta del archivo

# Funci√≥n para generar poes√≠a
def generar_poesia(tema):
    if len(tema.strip()) < 2:
        return "Por favor, ingresa una palabra o frase completa. Una sola letra no es suficiente.", None
    if not tema.strip():
        return "Por favor, ingresa un tema para generar la poes√≠a.", None

    try:
        # Solicitar poes√≠a
        prompt = (
            f"Escribe una poes√≠a creativa y √∫nica sobre el tema '{tema}'. "
            "Debe tener un lenguaje evocador y rimas naturales. "
            "Debe ser breve, con dos p√°rrafos como m√°ximo."
        )
        respuesta = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            max_tokens=300,
            temperature=0.7
        )
        poesia = respuesta["choices"][0]["message"]["content"].strip()

        # Generar narraci√≥n de la poes√≠a con gTTS
        narrador = gTTS(poesia, lang="es", tld="com")
        narracion_path = "/content/poesia_narrada.mp3"
        narrador.save(narracion_path)  # Guardar el archivo de la narraci√≥n

        # Verificar que el archivo de m√∫sica est√© en el directorio
        if not os.path.exists(musica_path):
            return "No se encuentra el archivo de m√∫sica. Aseg√∫rate de que la ruta sea correcta.", None

        # Combinar narraci√≥n y m√∫sica
        narracion_audio = AudioSegment.from_file(narracion_path)
        musica_audio = AudioSegment.from_file(musica_path).apply_gain(-15)  # Reducir volumen de la m√∫sica
        combinado = musica_audio.overlay(narracion_audio)


        # Guardar combinaci√≥n final
        combinado_path = "/content/poesia_con_musica.mp3"
        combinado.export(combinado_path, format="mp3")

        return poesia, combinado_path

    except Exception as e:
        return f"Error: {e}", None

# Interfaz de Gradio
with gr.Blocks() as interfaz:
    gr.Markdown("<h1>üåü Generador de Poes√≠as con Narraci√≥n y M√∫sica üåü</h1>")
    gr.Markdown("Escribe un tema y recibe una poes√≠a √∫nica narrada con m√∫sica de fondo. ¬°Int√©ntalo!")

    tema = gr.Textbox(
        lines=2,
        label="Tema de la poes√≠a üé≠",
        placeholder="Escribe aqu√≠ el tema sobre el que quieres la poes√≠a..."
    )

    boton = gr.Button("üé∂ Crear Poes√≠a Narrada üé∂")
    salida_texto = gr.Textbox(
        label="Poes√≠a Generada üìù",
        placeholder="Aqu√≠ aparecer√° tu poes√≠a. ¬°Espera un momento!"
    )
    salida_audio = gr.Audio(label="Poes√≠a Narrada con M√∫sica üéß")

    def procesar_poesia(tema):
        poesia, audio_path = generar_poesia(tema)
        if audio_path:
            return poesia, audio_path
        return poesia, None

    boton.click(procesar_poesia, inputs=tema, outputs=[salida_texto, salida_audio])

interfaz.launch()
```
