```python
!pip install openai gradio gtts pydub requests
!pip install openai==0.28.0

import requests
import openai
import gradio as gr
from gtts import gTTS
from pydub import AudioSegment
import os

# Configurar la API Key
openai.api_key = "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"

# Ruta al archivo de música de fondo que ya tienes
musica_path=   "/content/musica_fondo.mp3" # ruta del archivo

# Función para generar poesía
def generar_poesia(tema):
    if len(tema.strip()) < 2:
        return "Por favor, ingresa una palabra o frase completa. Una sola letra no es suficiente.", None
    if not tema.strip():
        return "Por favor, ingresa un tema para generar la poesía.", None

    try:
        # Solicitar poesía
        prompt = (
            f"Escribe una poesía creativa y única sobre el tema '{tema}'. "
            "Debe tener un lenguaje evocador y rimas naturales. "
            "Debe ser breve, con dos párrafos como máximo."
        )
        respuesta = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            max_tokens=300,
            temperature=0.7
        )
        poesia = respuesta["choices"][0]["message"]["content"].strip()

        # Generar narración de la poesía con gTTS
        narrador = gTTS(poesia, lang="es", tld="com")
        narracion_path = "/content/poesia_narrada.mp3"
        narrador.save(narracion_path)  # Guardar el archivo de la narración

        # Verificar que el archivo de música esté en el directorio
        if not os.path.exists(musica_path):
            return "No se encuentra el archivo de música. Asegúrate de que la ruta sea correcta.", None

        # Combinar narración y música
        narracion_audio = AudioSegment.from_file(narracion_path)
        musica_audio = AudioSegment.from_file(musica_path).apply_gain(-15)  # Reducir volumen de la música
        combinado = musica_audio.overlay(narracion_audio)


        # Guardar combinación final
        combinado_path = "/content/poesia_con_musica.mp3"
        combinado.export(combinado_path, format="mp3")

        return poesia, combinado_path

    except Exception as e:
        return f"Error: {e}", None

# Interfaz de Gradio
with gr.Blocks() as interfaz:
    gr.Markdown("<h1>🌟 Generador de Poesías con Narración y Música 🌟</h1>")
    gr.Markdown("Escribe un tema y recibe una poesía única narrada con música de fondo. ¡Inténtalo!")

    tema = gr.Textbox(
        lines=2,
        label="Tema de la poesía 🎭",
        placeholder="Escribe aquí el tema sobre el que quieres la poesía..."
    )

    boton = gr.Button("🎶 Crear Poesía Narrada 🎶")
    salida_texto = gr.Textbox(
        label="Poesía Generada 📝",
        placeholder="Aquí aparecerá tu poesía. ¡Espera un momento!"
    )
    salida_audio = gr.Audio(label="Poesía Narrada con Música 🎧")

    def procesar_poesia(tema):
        poesia, audio_path = generar_poesia(tema)
        if audio_path:
            return poesia, audio_path
        return poesia, None

    boton.click(procesar_poesia, inputs=tema, outputs=[salida_texto, salida_audio])

interfaz.launch()
```
